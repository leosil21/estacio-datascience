```python
import numpy as np
import pandas as pd
import random
import warnings
import matplotlib.pyplot as plt
import seaborn as sea
#from folium import plugins
import geopandas as gpd
import math
import pyarrow
from matplotlib.colors import LogNorm

warnings.filterwarnings("ignore")
%matplotlib inline

import statsmodels.formula.api as sm
from statsmodels.api import add_constant
```

```python
## carrega dataframe inicial
#df = pd.read_excel('./data/bnb/contratacoes/contratacoes-fne-2023-12.ods', engine="odf")
#df.to_parquet('./data/bnb/contratacoes/contratacoes-fne-2023-12.parquet')

# uma vez salvo em parquet, utilizar versão parquet
df = pd.read_parquet('./data/bnb/contratacoes/contratacoes-fne-2023-12.parquet', engine='fastparquet')
#df.head()
```

```python
## Carrega domínio de municípios
df_cod_mun = pd.read_excel('data/ibge/codigo-municipios.xlsx')
df_cod_mun['CD_IBGE'] = df_cod_mun['CD_IBGE'].apply(str).str[:7]
df_cod_mun.head()
```

```python
## Cria dataframe unindo a base inicial e a de municipios para buscar as chaves do IBGE
df_contratacoes = pd.merge(
    left=df, 
    right=df_cod_mun,
    how='left',
    left_on=['UF_MUN', 'NM_MUN'],
    right_on=['sg_uf', 'nm_mun'],
)

df_contratacoes["CD_IBGE"] = df_contratacoes["CD_IBGE"].apply(str).str[:7]
df_contratacoes.head()
```

```python
## carrega dataframe da população estimada por municipio
df_pop = pd.read_excel('data/ibge/populacao/POP2021_20240624.xlsx')
#df_pop = pd.read_csv('data/ibge/populacao/POP2022_Municipios_20230622.csv',sep=';')
df_pop["CD_MUN"] = df_pop["CD_MUN"].apply(str).str[:7]
df_pop["POPULACAO"] = df_pop["POPULAÇÃO ESTIMADA"]
df_pop.head()
```

```python
# Prepara os dados
# Carrega Malha de municipios
fp = 'data/ibge/mapas/BR_Municipios_2022/BR_Municipios_2022.shp'
map_df = gpd.read_file(fp)

# Carrega Malha de UF
fp_uf = 'data/ibge/mapas/BR_UF_2022/BR_UF_2022.shp'
map_uf_df = gpd.read_file(fp_uf)

# Cria dataframe de contratacoes por município
df_contratacoes_mun = pd.DataFrame(df_contratacoes, columns=['CD_IBGE','VALOR_CTR'])
df_contratacoes_uf = pd.DataFrame(df, columns=['UF_MUN','VALOR_CTR'])

df_contratacoes_mun = df_contratacoes_mun.groupby(['CD_IBGE'], as_index=False).sum()
df_contratacoes_uf = df_contratacoes_uf.groupby(['UF_MUN'], as_index=False).sum()
```

```python
## Principal função para plotagem do mapa
def geraHeatMapBrasil(titulo, variable, df_mun, df_uf, norm):
    fig, ax = plt.subplots(1, figsize=(16, 16))
    df_mun.plot(column=variable, cmap='YlOrBr', linewidth=0.2, ax=ax, edgecolor='0.9', norm=norm)
    df_uf.plot(column=variable, cmap='YlOrBr', linewidth=0.8, ax=ax, edgecolor='0.1', norm=norm)
    ax.axis('off')
    ax.set_title(titulo, fontdict={'fontsize': '25', 'fontweight' : '3'})
    ax.annotate('fonte: Contratações Curto-Longo Prazo 2023',xy=(0.1, 0.03), xycoords='figure fraction', horizontalalignment='left', verticalalignment='top', fontsize=12, color='#A6193C')
    sm = plt.cm.ScalarMappable(cmap='YlOrBr', norm=norm)
    sm._A = []
    fig.savefig('./img/'+titulo+'_map_export.png', dpi=300)
```

```python
## Processamento das informações de contratacoes totais por municipio

mapa_contratacoes_mun = map_df.set_index('CD_MUN').join(df_contratacoes_mun.set_index('CD_IBGE'))
mapa_contratacoes_uf = map_uf_df.set_index('SIGLA_UF').join(df_contratacoes_uf.set_index('UF_MUN'))
mapa_contratacoes_uf['VALOR_CTR'] = mapa_contratacoes_uf['VALOR_CTR'] * 0

variable = 'VALOR_CTR'
vmin, vmax = min(df_contratacoes_mun.VALOR_CTR), max(df_contratacoes_mun.VALOR_CTR)
log_norm = LogNorm(vmin=vmin, vmax=vmax)
geraHeatMapBrasil('Mapa de Calor - Contratações','VALOR_CTR', mapa_contratacoes_mun, mapa_contratacoes_uf, log_norm)
```

```python
## Processamento das informações de contratacoes per capita por municipio

df_pop_final = pd.DataFrame(df_pop, columns=['CD_MUN','POPULACAO'])
df_contratacoes_mun_percapita = df_pop_final.set_index('CD_MUN').join(df_contratacoes_mun.set_index('CD_IBGE'))
df_contratacoes_mun_percapita['VALOR_CTR'] = df_contratacoes_mun_percapita['VALOR_CTR'] / df_contratacoes_mun_percapita['POPULACAO']

df_contratacoes_mun_percapita = df_contratacoes_mun_percapita.reset_index()
df_contratacoes_mun_percapita = df_contratacoes_mun_percapita.loc[df_contratacoes_mun_percapita['VALOR_CTR'] > 1]
```

```python
## Plotagem das informações de contratacoes per capita por municipio

mapa_contratacoes_percapita = map_df.set_index('CD_MUN').join(df_contratacoes_mun_percapita.set_index('CD_MUN'))
mapa_contratacoes_uf = map_uf_df.set_index('SIGLA_UF').join(df_contratacoes_uf.set_index('UF_MUN'))
mapa_contratacoes_uf['VALOR_CTR'] = mapa_contratacoes_uf['VALOR_CTR'] * 0

vmin, vmax = min(df_contratacoes_mun_percapita.VALOR_CTR), max(df_contratacoes_mun_percapita.VALOR_CTR)
log_norm=LogNorm(vmin=vmin, vmax=vmax)
geraHeatMapBrasil('Mapa de Calor - Contratações per capita','VALOR_CTR', mapa_contratacoes_percapita, mapa_contratacoes_uf, log_norm)
```

```python
## Carregando inforamções do PIB

df_pib = pd.read_excel('data/ibge/pib/PIB dos Munic¡pios - base de dados 2010-2021.xlsx')
df_pib['PIB'] = df_pib['Produto Interno Bruto, \na preços correntes\n(R$ 1.000)']
df_pib['PIB_percapita'] = df_pib['Produto Interno Bruto per capita, \na preços correntes\n(R$ 1,00)']
```

```python
## Processamento das informações de contratacoes per capita por municipio

df_pib_final = pd.DataFrame(df_pib[df_pib['Ano'] == max(df_pib['Ano'])], columns=['Código do Município','PIB', 'Ano'])
df_pib_final['Código do Município'] = df_pib_final['Código do Município'].apply(str)

df_contratacoes_mun_pib = df_pib_final.set_index('Código do Município').join(df_contratacoes_mun.set_index('CD_IBGE'))
df_contratacoes_mun_pib = df_contratacoes_mun_pib.dropna()
df_contratacoes_mun_pib['VALOR_CTR'] = df_contratacoes_mun_pib['VALOR_CTR'] / df_contratacoes_mun_pib['PIB']

df_contratacoes_mun_pib = df_contratacoes_mun_pib.reset_index()
df_contratacoes_mun_pib = df_contratacoes_mun_pib.loc[df_contratacoes_mun_pib['VALOR_CTR'] > 1]
```

```python
## Plotagem das informações de contratacoes per capita por municipio
mapa_contratacoes_pib = map_df.set_index('CD_MUN').join(df_contratacoes_mun_pib.set_index('Código do Município'))
vmin, vmax = min(df_contratacoes_mun_pib.VALOR_CTR), max(df_contratacoes_mun_pib.VALOR_CTR)
log_norm=LogNorm(vmin=vmin, vmax=vmax)
geraHeatMapBrasil('Mapa de Calor - Contratações pelo PIB','VALOR_CTR', mapa_contratacoes_pib, mapa_contratacoes_uf, log_norm)
```

```python

mapa_contratacoes_mun_menores = mapa_contratacoes_mun.sort_values(by=['VALOR_CTR'])[:50]
vmin, vmax = min(mapa_contratacoes_mun_menores.VALOR_CTR), max(mapa_contratacoes_mun_menores.VALOR_CTR)
log_norm=LogNorm(vmin=vmin, vmax=vmax)
geraHeatMapBrasil('Mapa de Calor - 50 Menores Contratações pelo PIB','VALOR_CTR', mapa_contratacoes_mun_menores, mapa_contratacoes_uf, log_norm)
```

```python

mapa_contratacoes_mun_maiores = mapa_contratacoes_mun.sort_values(by=['VALOR_CTR'], ascending=False)[:50]
vmin, vmax = min(mapa_contratacoes_mun_maiores.VALOR_CTR), max(mapa_contratacoes_mun_maiores.VALOR_CTR)
log_norm=LogNorm(vmin=vmin, vmax=vmax)
geraHeatMapBrasil('Mapa de Calor - 50 Maiores Contratações pelo PIB','VALOR_CTR', mapa_contratacoes_mun_maiores, mapa_contratacoes_uf, log_norm)
```

```python
df_menores_contracoes_percapita = df_contratacoes_mun_percapita.sort_values(by=['VALOR_CTR'])[:10]

df_menores_contracoes_percapita = pd.merge(
    left=df_menores_contracoes_percapita, 
    right=df_cod_mun,
    how='left',
    left_on=['CD_MUN'],
    right_on=['CD_IBGE'],
)
```

```python
df_idh = pd.read_excel('data/ibge/idh/IDH-M-rep1.xlsx')
df_idh["Municipio"] = df_idh["Municipio"].str.upper()
df_idh["Municipio"] = df_idh["Municipio"].str.normalize('NFKD').str.encode('ascii', errors='ignore').str.decode('utf-8').apply(str).str.strip()
df_idh["Estado"] = df_idh["Estado"].apply(str).str.strip()

df_cod_mun['sg_uf'] = df_cod_mun['sg_uf'].apply(str).str.strip()
df_cod_mun['nm_mun'] =  df_cod_mun['nm_mun'].apply(str).str.strip()
```

```python
df_idh_mun = pd.merge(
    left=df_idh, 
    right=df_cod_mun,
    how='left',
    left_on=['Estado', 'Municipio'],
    right_on=['sg_uf', 'nm_mun']
)
```

```python
df_mun_final = pd.DataFrame(df_idh_mun, columns=['sg_uf','nm_mun', 'CD_IBGE', 'IDHM 1991','IDHM 2000', 'IDHM 2010'])
df_mun_final = df_mun_final.dropna()
```

```python
df_contratacoes_gp_mun = df_contratacoes[['CD_IBGE','VALOR_CTR']].groupby(['CD_IBGE'], as_index=False).sum()
df_contratacoes_gp_mun_setor = df_contratacoes[['CD_IBGE','SETOR','VALOR_CTR']].groupby(['CD_IBGE', 'SETOR'], as_index=False).sum()
```

```python
for setor in df_contratacoes_gp_mun_setor['SETOR'].unique():
    df_mun_final[setor] = [0.0 for i in range(len(df_mun_final))]
    for index, row in df_mun_final.iterrows():
        df_mun_final.loc[index, setor] = sum(df_contratacoes_gp_mun_setor[(df_contratacoes_gp_mun_setor['SETOR'] == setor) & (df_contratacoes_gp_mun_setor['CD_IBGE'] == row['CD_IBGE'])]['VALOR_CTR']) #sum(df_contratacoes_gp_mun_setor[(df_contratacoes_gp_mun_setor['SETOR'] == setor) & (df_contratacoes_gp_mun_setor['CD_IBGE'] == row['CD_IBGE'])]['VALOR_CTR'])
```

```python
for index, row in df_mun_final.iterrows():
    df_mun_final.loc[index, 'Valor Contratacao'] = sum(df_contratacoes_gp_mun_setor[(df_contratacoes_gp_mun_setor['CD_IBGE'] == row['CD_IBGE'])]['VALOR_CTR'])
    df_mun_final.loc[index, 'PIB'] = sum(df_contratacoes_mun_pib[(df_contratacoes_mun_pib['Código do Município'] == row['CD_IBGE'])]['PIB'])
    df_mun_final.loc[index, 'Valor Contratacao ponderado pelo PIB'] = sum(df_contratacoes_mun_pib[(df_contratacoes_mun_pib['Código do Município'] == row['CD_IBGE'])]['VALOR_CTR'])
    df_mun_final.loc[index, 'População Estimada'] = sum(df_contratacoes_mun_percapita[(df_contratacoes_mun_percapita['CD_MUN'] == row['CD_IBGE'])]['POPULACAO'])
    df_mun_final.loc[index, 'Valor Contratacao per capita'] = sum(df_contratacoes_mun_percapita[(df_contratacoes_mun_percapita['CD_MUN'] == row['CD_IBGE'])]['VALOR_CTR'])

#df_contratacoes_mun_pib.head()
df_mun_final[df_mun_final['Valor Contratacao'] > 0].head()
```
